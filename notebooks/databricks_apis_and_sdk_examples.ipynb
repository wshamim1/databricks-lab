# Databricks APIs and SDK Examples

This notebook summarizes common Databricks automation patterns:

- REST API authentication basics
- job creation payloads
- cluster creation payloads
- Python SDK client setup
- SDK examples for jobs and clusters

## REST API environment variables

Databricks REST calls typically use a workspace host URL and an access token.

```python
rest_env_example = {
    "DATABRICKS_HOST": "https://<your-workspace-url>",
    "DATABRICKS_TOKEN": "<personal-access-token>"
}

rest_env_example
```

## Job creation payload

This is the kind of JSON payload sent to the jobs create endpoint.

```python
job_payload = {
    "name": "orders_daily_pipeline",
    "max_concurrent_runs": 1,
    "tasks": [
        {
            "task_key": "ingest_orders",
            "notebook_task": {
                "notebook_path": "/Workspace/Shared/notebooks/ingest_orders",
                "base_parameters": {
                    "run_date": "2026-04-04",
                    "run_mode": "incremental"
                }
            },
            "new_cluster": {
                "spark_version": "15.4.x-scala2.12",
                "node_type_id": "Standard_DS3_v2",
                "num_workers": 2
            }
        }
    ]
}

job_payload
```

## Cluster creation payload

Cluster creation can also be handled directly through REST requests.

```python
cluster_payload = {
    "cluster_name": "demo-api-cluster",
    "spark_version": "15.4.x-scala2.12",
    "node_type_id": "Standard_DS3_v2",
    "autotermination_minutes": 20,
    "num_workers": 2
}

cluster_payload
```

## Python SDK setup

The Databricks SDK for Python provides a higher-level interface than raw REST calls.

```python
sdk_setup_example = """
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()
"""

print(sdk_setup_example)
```

## SDK job creation example

This example mirrors the REST payload using the Python SDK.

```python
sdk_job_example = """
job = w.jobs.create(
    name=\"orders_daily_pipeline\",
    max_concurrent_runs=1,
    tasks=[
        {
            \"task_key\": \"ingest_orders\",
            \"notebook_task\": {
                \"notebook_path\": \"/Workspace/Shared/notebooks/ingest_orders\"
            },
            \"new_cluster\": {
                \"spark_version\": \"15.4.x-scala2.12\",
                \"node_type_id\": \"Standard_DS3_v2\",
                \"num_workers\": 2
            }
        }
    ]
)
"""

print(sdk_job_example)
```
